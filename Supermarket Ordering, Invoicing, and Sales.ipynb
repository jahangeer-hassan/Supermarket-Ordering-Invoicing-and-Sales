#  Supermarket Ordering, Invoicing, and Sales

#  Introduction

Exploratory Data Analysis (EDA) is an important process in data science used to understand datasets, identify patterns, clean data, and extract meaningful business insights. In this project, three different business-related datasets — **Invoices**, **OrderLeads**, and **SalesTeam** — were analyzed using Python libraries such as Pandas, Matplotlib, and Seaborn.

The main objective of this project was to explore customer ordering behavior, sales lead performance, and sales team management using statistical analysis and visualizations. Multiple EDA techniques were applied, including:

* Data cleaning
* Missing value analysis
* Duplicate checking
* Statistical summaries
* Trend analysis
* Correlation analysis
* Data visualization

Through this project, raw business data was transformed into useful insights that can support smarter business decisions and operational improvements.


# Dataset Overview

## 1. Invoices Dataset

The Invoices dataset contains information related to:

* Company names
* Meal prices

## What is the distribution of meal prices?

```
plt.figure(figsize=(8,5))
sns.histplot(df['Meal Price'], bins=30, kde=True)
plt.title("Distribution of Meal Price")
plt.xlabel("Meal Price")
plt.ylabel("Frequency")
plt.show()
```

## Which meal type has the highest count?

```
plt.figure(figsize=(7,5))
sns.countplot(x='Type of Meal', data=df)
plt.title("Count of Meal Types")
plt.show()
```

## What is the average meal price for each meal type?

```
plt.figure(figsize=(7,5))
sns.barplot(x='Type of Meal', y='Meal Price', data=df)
plt.title("Average Meal Price by Meal Type")
plt.show()
```

## How has the number of orders changed over the years?

```
df['Year'] = df['Date'].dt.year

plt.figure(figsize=(8,5))
df['Year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Orders Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Orders")
plt.show()
```

## Which are the top 10 companies with the most orders?

```
top_companies = df['Company Id'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_companies.plot(kind='bar')
plt.title("Top 10 Companies by Orders")
plt.xlabel("Company ID")
plt.ylabel("Orders")
plt.show()
```

## What does the correlation heatmap show?

```
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```
plt.figure(figsize=(8,5))
sns.histplot(df['Order Value'], bins=30, kde=True)
plt.title("Order Value Distribution")
plt.show()
```

## What is the ratio of converted and non-converted orders?

```
plt.figure(figsize=(6,4))
sns.countplot(x='Converted', data=df)
plt.title("Converted vs Non-Converted")
plt.show()
```

## What is the average order value of converted orders?

```
plt.figure(figsize=(6,4))
sns.barplot(x='Converted', y='Order Value', data=df)
plt.title("Average Order Value by Conversion")
plt.show()
```

## How have orders changed over the years?

```
df['Year'] = df['Date'].dt.year

plt.figure(figsize=(8,5))
df['Year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Orders by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

## Which companies have the highest number of orders?

```
top_companies = df['Company Name'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_companies.plot(kind='bar')
plt.title("Top 10 Companies")
plt.xlabel("Company Name")
plt.ylabel("Orders")
plt.show()
```

## What does the correlation heatmap show?

```
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(5,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='viridis')
plt.title("Correlation Heatmap")
plt.show()
```

```
print("\nUnique Sales Reps:", df['Sales Rep'].nunique())
print("Unique Companies:", df['Company Name'].nunique())


```

## Which sales representatives manage the highest number of companies?

```
top_reps = df['Sales Rep'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_reps.plot(kind='bar')
plt.title("Top 10 Sales Representatives")
plt.xlabel("Sales Rep")
plt.ylabel("Number of Companies")
plt.show()
```

## Which companies appear most frequently in the dataset?

```
top_companies = df['Company Name'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_companies.plot(kind='bar')
plt.title("Top Companies")
plt.xlabel("Company")
plt.ylabel("Frequency")
plt.show()
```

## What is the distribution of sales representatives in the dataset?

```
plt.figure(figsize=(8,5))
sns.countplot(y='Sales Rep', data=df,
              order=df['Sales Rep'].value_counts().head(15).index)
plt.title("Sales Rep Distribution")
plt.show()
```

## How many companies are managed by each sales representative?

```
company_per_rep = df.groupby('Sales Rep')['Company Id'].count().sort_values(ascending=False)

plt.figure(figsize=(10,5))
company_per_rep.head(10).plot(kind='bar')
plt.title("Companies Managed by Sales Reps")
plt.xlabel("Sales Rep")
plt.ylabel("Companies Count")
plt.show()
```
